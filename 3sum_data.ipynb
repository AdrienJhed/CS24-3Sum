```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
points = np.array([100, 200, 400, 800, 1600, 4000])
time = np.array([1.441, 8.895, 71.57, 538.197, 4242.3, 73129.7])
```

```python
plt.style.use('_mpl-gallery')

plt.figure(figsize=(8, 6))
plt.plot(points, time, marker = 'o', color = 'k', markerfacecolor = 'r')
plt.xlabel("n (size of input array)")
plt.ylabel("T(n) (time (ms))")
plt.title("3Sum Analysis")


plt.show()
```

```python
plt.style.use('_mpl-gallery')

log2_time = np.log2(time)
log2_points = np.log2(points)


plt.figure(figsize=(8, 6))
plt.plot(log2_points, log2_time, marker = 'o', color = 'k', markerfacecolor = 'r')
plt.xlabel("Log2(n) (size of input array)")
plt.ylabel("Log2(T(n)) (time (ms))")
plt.title("3Sum Analysis")


plt.show()
```

```python
log2_time = np.log2(time)
log2_points = np.log2(points)

# Fit a line: log(T(n)) = m * n + b
m, b = np.polyfit(log2_points, log2_time, 1)

# Sort points for smooth line plotting
sorted_indices = np.argsort(points)
sorted_points = np.array(points)[sorted_indices]
fit_line = m * sorted_points + b

# Plotting
plt.figure(figsize=(8, 6))

plt.scatter(log2_points, log2_time, color='blue', label='Data')
plt.plot(sorted_points, fit_line, color='red', label=f'Fit: slope = {m:.2f}')
plt.xlabel('Input Size Log(n)')
plt.ylabel('log(T(n))')
plt.title('Log of Running Time vs. Log Input Size')
plt.grid(True)
plt.legend()
plt.show()

# Print slope and base
print(f"Slope of the fitted line: {m:.2f}")
print(f"Base b = 2^slope  {2**m:.2f}")
```

```python
log2_time = np.log2(time)
log2_points = np.log2(points)

# Fit a line: log(T(n)) = m * log(n) + b
m, b = np.polyfit(log2_points, log2_time, 1)

# Generate line for fitted values
x_fit = np.linspace(min(log2_points), max(log2_points), 100)
y_fit = m * x_fit + b

# Set style
plt.style.use('_mpl-gallery')

# Plotting
plt.figure(figsize=(8, 6))

# Plot data points
plt.plot(log2_points, log2_time, 'o', color='k', markerfacecolor='r', label='Data')

# Plot regression line
plt.plot(x_fit, y_fit, color='blue', label=f'Fit: slope = {m:.2f}')

# Labels and title
plt.xlabel("Log(n) (size of input array)")
plt.ylabel("Log(T(n)) (time in ms)")
plt.title("3Sum Analysis: Log-Time vs. Log-Input Size")

plt.grid(True)
plt.legend()
plt.show()

# Print the slope and estimated base
print(f"Slope of the fitted line: {m:.2f}")
print(f"Estimated complexity base  2^{m:.2f}  {2**m:.2f}")
```

```python
coeffs = np.polyfit(points, time, 3)
poly = np.poly1d(coeffs)

# Generate smooth curve
x_fit = np.linspace(min(points), max(points), 200)
y_fit = poly(x_fit)

# Set style
plt.style.use('_mpl-gallery')

# Plotting
plt.figure(figsize=(8, 6))

# Plot data points
plt.plot(points, time, 'o', color='k', markerfacecolor='r', label='Data')

# Plot fitted curve
plt.plot(x_fit, y_fit, color='blue', label='Fitted Curve')

# Labels and title
plt.xlabel("n (size of input array)")
plt.ylabel("T(n) (time in ms)")
plt.title("3Sum Analysis: Time vs. Input Size")

plt.grid(True)
plt.legend()
plt.show()
```
